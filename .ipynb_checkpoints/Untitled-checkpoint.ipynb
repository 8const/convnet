```python
from matplotlib.pyplot import imshow
import numpy as np
import matplotlib.pyplot as plt
from cv2 import imread, IMREAD_GRAYSCALE
```

```python
pixel = lambda local, kernel: (local * kernel).sum()
get_local = lambda image, i, j: image[i:i+5, j:j+5]

#boring people know it's actually cross-correlation
def convolution(img, kernel):
    result = np.zeros_like(img)[4:, 4:]
    for i in range(0, result.shape[0]):
        for j in range(0, result.shape[1]):
            p = pixel(get_local(img, i, j), kernel)
            result[i,j] = p
    return result
```

```python
Gx = convolution(img, kernel)
Gy = convolution(img, kernel.T)
G = np.sqrt(np.square(Gx/255) + np.square(Gy/255))
```

```python
#gets data for training & testing;
def fetch(url):
    import requests, gzip, os, hashlib, numpy
    fp = os.path.join("/tmp", hashlib.md5(url.encode('utf-8')).hexdigest())
    if os.path.isfile(fp):
        with open(fp, "rb") as f:
            dat = f.read()
    else:
        with open(fp, "wb") as f:
            dat = requests.get(url).content
            f.write(dat)
    return numpy.frombuffer(gzip.decompress(dat), dtype=np.uint8).copy()


X_train = fetch("http://yann.lecun.com/exdb/mnist/train-images-idx3-ubyte.gz")[0x10:].reshape((-1, 28, 28))
Y_train = fetch("http://yann.lecun.com/exdb/mnist/train-labels-idx1-ubyte.gz")[8:]
X_test = fetch("http://yann.lecun.com/exdb/mnist/t10k-images-idx3-ubyte.gz")[0x10:].reshape((-1, 28, 28))
Y_test = fetch("http://yann.lecun.com/exdb/mnist/t10k-labels-idx1-ubyte.gz")[8:]
```

```python
#gets average over 2x2 chanks and puts into 
#image twice as small as the input
#not even-shaped inputs might not work well
def avg_pool(img):
    h = int(img.shape[0]/2)
    w = int(img.shape[1]/2)
    result = np.zeros((h, w))
    chunk = lambda image, i, j: image[i:i+2, j:j+2]
    avg = lambda x: x.sum() / 4

    for i in range(h):
        for j in range(w):
            result[i,j] = avg(chunk(img, 2*i, 2*j))
    return result
imshow(avg_pool(x))
avg_pool(x).shape
```

```python
def convolution_8(img, kernel):
    pixel = lambda local, kernel: (local * kernel).sum()
    get_local = lambda image, i, j: image[i:i+8, j:j+8]

    result = np.zeros_like(img)[7:, 7:]
    for i in range(0, result.shape[0]):
        for j in range(0, result.shape[1]):
            p = pixel(get_local(img, i, j), kernel)
            result[i,j] = p
    return result

def deltaC2sig(q):
    res = np.zeros((8,8))
    for i in range(8):
        for j in range(8):
            res[i, j] = dC2sig(q, i, j)
    return res

def dk2(p, q):
    return convolution_8(np.rot90(np.rot90(s1[p])), deltaC2sig(q))
```

```python
def prod_of_inters(x, y, i, j):
    return (x[i:8, j:8] * y[0:8-i, 0:8-j]).sum()
```

```python
def interval(j):
    if j < 4:
        print("TOO SMOL")
        return
    if j-4 <= 4:
        return range(4, j+1)
    if j > 11:
        return range(j-4, j+1)
    else:
        return range(j-4, j+1)
    print("KEK PIZDA")
```

```python
def intersect(r, rr):
    return range(max(r.start, rr.start), min(r.stop, rr.stop))
```

```python
def cords_in_kernel(j):
    if j > 15 or j < 4:
        print("BAD j IN cords_in_kernql(j)")
        raise Exception
    if len(interval(j)) == 0:
        print("NOTHING")
    if len(interval(j)) == 5:
        return range(0, 5)
    if j > 11:
        return range(0, len(interval(j)))
    if j < 8:
        return range(5-len(interval(j)), 5)
    print("SOOQUAAA")
```
